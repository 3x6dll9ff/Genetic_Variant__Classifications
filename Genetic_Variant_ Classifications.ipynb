```python
import pandas as pd

df = pd.read_csv('clinvar_conflicting.csv', low_memory=False)
df.columns.tolist()
```

```python
df['CLNSIGINCL'].value_counts()
```

```python

def simplify_clnsig(x):
    x = str(x)
    parts = x.split(':')
    label = parts[-1]  
    label = label.split('|')[0]  
    label = label.strip().lower() 

    if label in ['benign']:
        return 0
    elif label in ['pathogenic', 'likely_pathogenic']:
        return 1
    else:
        return -1 


df['label'] = df['CLNSIGINCL'].apply(simplify_clnsig)


df_small = df[df['label'] != -1]


df_small[['REF','ALT','label']].head()
```
